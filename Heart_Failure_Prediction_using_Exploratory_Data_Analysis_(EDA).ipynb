```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

```

```python
dataset = pd.read_csv('heart.csv')
```

```python
dataset.columns
```

```python
X = dataset.drop(columns='HeartDisease') # All attributes expect HeartDisease

Y = dataset['HeartDisease'] # Store Dropped column which is Class_name
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,stratify=Y,random_state = 2)
print(X.shape,X_train.shape,X_test.shape)
```

# Decision Tree Classifier

```python
DTC = DecisionTreeClassifier(criterion='gini',splitter='best',
                             max_depth=None,min_samples_split=2, min_samples_leaf=1,
                             min_weight_fraction_leaf=0)
DTC.fit(X_train,Y_train)

X_train_prediction = DTC.predict(X_train)
training_data = accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score of training_data: ',training_data)

X_test_prediction = DTC.predict(X_test)
test_data = accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score of test_data: ',test_data)
print()
training_data = precision_score(X_train_prediction,Y_train)
print('Precision Score of training_data: ',training_data)

test_data = precision_score(X_test_prediction,Y_test)
print('Precision Score of test_data: ',test_data)
print()
training_data = f1_score(X_train_prediction,Y_train)
print('f1 Score of Trainig_data: ',training_data)

test_data = f1_score(X_test_prediction,Y_test)
print('f1 Score of Test_data: ',test_data)
```

# Logistic Regression

```python
LR = LogisticRegression(random_state=None, max_iter=100, penalty='l2')
LR.fit(X_train,Y_train)

X_train_prediction = LR.predict(X_train)
training_data = accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score of training_data: ',training_data)

X_test_prediction = LR.predict(X_test)
test_data = accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score of test_data: ',test_data)
print()
training_data = precision_score(X_train_prediction,Y_train)
print('Precision Score of training_data: ',training_data)

test_data = precision_score(X_test_prediction,Y_test)
print('Precision Score of test_data: ',test_data)
print()
training_data = f1_score(X_train_prediction,Y_train)
print('f1 Score of Trainig_data: ',training_data)

test_data = f1_score(X_test_prediction,Y_test)
print('f1 Score of Trainig_data: ',test_data)
```

# KNearest Neighbors

```python
KNN = KNeighborsClassifier(n_neighbors=7, weights='uniform', algorithm='auto', leaf_size=30)
KNN.fit(X_train,Y_train)

X_train_prediction = KNN.predict(X_train)
training_data = accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score of training_data: ',training_data)

X_test_prediction = KNN.predict(X_test)
test_data = accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score of test_data: ',test_data)
print()
training_data = precision_score(X_train_prediction,Y_train)
print('Precision Score of training_data: ',training_data)

test_data = precision_score(X_test_prediction,Y_test)
print('Precision Score of test_data: ',test_data)
print()
training_data = f1_score(X_train_prediction,Y_train)
print('f1 Score of Trainig_data: ',training_data)

test_data = f1_score(X_test_prediction,Y_test)
print('f1 Score of Trainig_data: ',test_data)
```
